# Descrição do Dataset
Link: https://www.kaggle.com/austinreese/craigslist-carstrucks-data

O dataset contém os carros usados e à venda listados no Craiglist dentro dos Estados Unidos. Dados foram coletados através de um scrapper e atualizados constantemente dentro do período de alguns meses.
O dataset contém quase todas as informações relevantes mostradas no Craiglist.

O dataset contém 435.849 amostras.


## Colunas do Dataset

| Coluna                 | Descrição                                                                                    | Tipo         |
|------------------------|----------------------------------------------------------------------------------------------|--------------|
| id                     | Identificador da entrada, número único que representa o anúncio que foi postado no Craiglist | Integer      |
| url                    | Url do anúncio                                                                               | String       |
| region                 | Região dos Estados Unidos no Craiglist                                                       | String       |
| price                  | Preço (em dólar) do carro no anúncio                                                         | Float        |
| year                   | Ano de fabricaçao do veiculo                                                                 | Integer      |
| manufacturer           | Fabricante do veículo                                                                        | String       |
| model                  | Modelo/nome do veículo                                                                       | String       |
| condition              | Condição do veículo (boa, ruim, seminovo, excelente...)                                      | String       |
| cylinders              | Número de cilindros                                                                          | Integer      |
| fuel                   | Tipo de combustivel                                                                          | String       |
| odometer               | Milhas viajada pelo veículo                                                                  | Float        |
| title_status           | O status do veículo (limpo, reconstruído, apenas partes...)                                  | String       |
| transmission           | Tipo de transmissão do veículo                                                               | String       |
| vin                    | Número de identificação do veículo                                                           | String       |
| drive                  | Tipo de tração: rwd (traseira), 4wd (quatro rodas) e fwd (dianteira)                         | String       |
| size                   | Tamanho do veículo                                                                           | String       |
| type                   | Tipo genérico do veículo (pickup, offroad, hatchback, sedan, suv...)                         | String       |
| paint_color            | Cor do veículo                                                                               | String       |
| description            | Descrição do veículo no anúncio                                                              | String       |
| state                  | Estado onde foi feito o anúncio                                                              | String       |
| lat                    | Latitude do anúncio                                                                          | Float        |
| long                   | Longitude do anúncio                                                                         | Float        |
| region_url (removida)  | Url da região do Estados Unidos                                                              | String       |
| county     (removida)  | Coluna inútil deixada por engano, será desconsiderada durante o projeto                      | Float        |
| image_url  (removida)  | Url da imagem                                                                                | String       |

### Visualização de dados
#### Preço (price)
De acordo com as análises abaixo pode-se notar que há vários anúncios o com preços de carro menor que U\$100, que são provavelmente outliers. 

Esses outliers foram retirados, resultando em uma redução de 9\% no tamanho do dataset, passando a ter 397.012 amostras.

```python
df['log_price'] = df['price'].apply(np.log10)
```

```python
df[df['price'] > 0]['price'].apply(np.log10).hist()
```

```python
df[df['log_price'] > 0].boxplot(column=['log_price'])
```

```python
too_low_price_df = df[df['price'] < 500]
too_low_price_df['price'].hist()
```

```python
too_high_price_df = df[df['price'] > 200000]
too_high_price_df['price'].hist()
```

```python
# Calculando o total de possíveis outliers de acordo com o preço
outliers = len(too_low_price_df) + len(too_high_price_df)
outliers

# Calculando a quantidade de anúncios do dataset ao retirar os outliers de preço
percent = (1 - (outliers / len(df)))
offers = len(df) * percent
offers
```

```python
filtered_df = df[(df['price'] >= 500) & (df['price'] <= 200000)]
filtered_df
```

```python
filtered_df['log_price'].hist()
```

```python
filtered_df.boxplot(column=['price'])
```

```python
filtered_df.drop(columns=['log_price'], inplace=True)
filtered_df.to_csv("datasets/no_price_outliers_vehicles.csv", index=False)
```
